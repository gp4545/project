```python
from random import randint
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

x=randint(0,50)
guess=-1
count=0
while guess!=x:
    guess=int(input('guess the number between 0 and 50: '))
    count =count+1
    if guess == x:
        score=1100-count*50
        print(f'congrats u read my mind in {count} attempts. you have scored {1100-count*50} point')
        break 
    elif guess < x:
        print("please guess a larger number")
    else:
        print("please guess smaller number")

data={
    'Targetnumber':[23,45,6,8,14,18,16],
    'totalattempts':[2,5,3,5,4,9,4],
    'point':[1000,850,900,450,800,400,500],
    'guess':[23,25,44,6,5,13,4,]}
df=pd.DataFrame(data)
df.to_csv("game.csv",index=False,header=False)
sns.lineplot(x=df.index,y='point',marker='o',data=df,)
plt.xlabel("game no")
plt.ylabel("score")
plt.title("Score in each game")
plt.show()

```
