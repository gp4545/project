```python
conn=sqlite3.connect('project.db')
c=conn.cursor()
```

```python
c.execute(
    """CREATE TABLE pro1('S.No.'int, 'Industry' text, 'No. of Companies' int, 'Total Market Cap.'int,
       'Avg Market Cap.'int, 'Median 'int, 'Avg Sales Growth'int, 'Avg OPM'int,
       'Avg ROCE'int, '1Y Return'int);""")
```

```python
def rsql(query):
    return pd.read_sql_query(query,conn)
```

```python
df.to_sql('pro1',conn,if_exists="replace",index=False)
```

```python
rsql('SELECT*FROM pro1 WHERE "Avg Sales Growth">10')
```

```python
rsql("""SELECT "Industry", SUM("Total Market Cap.")
FROM pro1 GROUP BY "Industry";""")
```

```python
rsql("""SELECT * FROM pro1 
ORDER BY "Avg Market Cap." DESC;""")
```

```python
rsql("""SELECT * FROM pro1 
LIMIT 5;""")
```

```python
rsql("""SELECT "Industry" AS Sector, "1Y Return" AS Annual_Return FROM pro1;""")
```

```python
rsql("""SELECT "Industry", COUNT(*) AS rows FROM pro1 GROUP BY "Industry" HAVING COUNT(*) > 1;""")
```

```python
rsql("""SELECT * FROM pro1 WHERE "Avg Market Cap." BETWEEN 10000 AND 50000;""")
```

```python
rsql("""SELECT AVG("Avg Market Cap.") AS avgcap, SUM("Total Market Cap.") AS totalcap, COUNT(*) AS totalrows FROM pro1;""")
```

```python
rsql("""SELECT DISTINCT "Industry" FROM pro1;""")
```
